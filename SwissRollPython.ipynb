```python
from sklearn.datasets import make_swiss_roll
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Use the make_swiss_roll function to generate a dataset
x, t = make_swiss_roll(n_samples=5000)
# The printed array is in x, z, y order
print(x)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
#ax.plot3D(x[0], x[1], x[2], 'gray')
#ax = fig.add_subplot(211, projection="3d")
ax.scatter(x[:, 0], x[:, 2], x[:, 1], c=t, cmap=plt.cm.Spectral)
```
